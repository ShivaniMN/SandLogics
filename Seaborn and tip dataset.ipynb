```python
tips
```

```python
#using subset of data for demonstration
df = tips.loc[:, ['total_bill','size','tip']]
```

```python
df.loc[0:20,'size'] = np.nan
df.loc[210:240,'total_bill'] = np.nan
```

```python
#mode imputation for size
mode_size = df['size'].mode()
df_modesize = df.fillna(value = {'size':mode_size[0]})
```

```python
df_modesize
```

```python
#Regression imputation on total_bills and check for correlation among variables
corr = df.corr()
corr.style.background_gradient().set_precision(2)
```

```python
#creating a subset of data where there are no missing values in the tip and also total_bill
df_bill_tip = df.dropna(axis=0,subset=['total_bill','tip'])
df_bill_tip = df_bill_tip.loc[:,['total_bill','tip']]
```

```python
df_bill_tip
```

```python
#total bill missing entries
missing_bill = df['total_bill'].isnull()

#tips of observation with toral_bill missing
tip_missing = pd.DataFrame(df['tip'][missing_bill])
```
