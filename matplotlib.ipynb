```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('predictive analysis')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('y label')
axes.set_title('Set Title')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)


axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')


axes[1].plot(y,x)
axes[1].set_title('Second plot')
axes[1].set_label('x')
axes[1].set_label('y')
```

```python
#Figure Size and DPI
```

```python
fig = plt.figure(figsize=(5,3))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='X,Y')
ax.plot(y,x,label='Y,X')
ax.plot(x,x**2,label='X square')
ax.plot(x,x**3,label='X cube')

ax.legend(loc=0)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(5,5))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_pic.png',dpi=200)
```
